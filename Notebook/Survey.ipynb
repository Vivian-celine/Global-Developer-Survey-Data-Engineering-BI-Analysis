```python
id_vars = [
    'Respondent_id', 'Year', 'Age', 'Gender', 'Country', 'Education Level',
    'Current Role', 'Experience years', 'Company Size', 'Yearly compensation',
    'Data Analytics Tool']

question_blocks = {
    'Programming Language': list(df.columns[8:19]),   # Python to Other Programming languages
    'IDE': list(df.columns[20:29]),                   # Jupyter to Other IDE's
    'Notebooks': list(df.columns[30:42]),             # Kaggle to Other Notebooks
    'Visualization Tools': list(df.columns[43:53]),   # Matplotlib to Other Visualization tools
    'ML Frameworks': list(df.columns[54:68]),         # Scikit-learn to Other ML Frameworks
    'Cloud Computing': list(df.columns[71:81]),       # Amazon Web to Other platforms
    'ML Products': list(df.columns[82:84]),           # Amazon Sagemaker to Other Machine Learning
    'Data Products': list(df.columns[85:98]),         # MYSQL to Other Data Products
    'BI Tools': list(df.columns[99:111]),             # Amazon QuickSight to Other Bi tools
    'Public Sharing Platforms': list(df.columns[112:120]) }  # Plotly Dash to Others
```

```python
melted_list = []

for question, cols in question_blocks.items():
    melted = df.melt(
        id_vars=id_vars,
        value_vars=cols,
        var_name='Option',
        value_name='Value')
    
    melted['Question'] = question 
    melted_list.append(melted)


DF_long = pd.concat(melted_list, ignore_index=True)
```

```python
DF_long_clean = DF_long.dropna(subset=['Value'])
```

```python
id_vars = ['Respondent_id', 'Gender', 'Education Level']

question_blocks = {
    'Programming Language': ['Python', 'R', 'SQL', 'C', 'C++', 'Java', 'Javascript', 'Julia', 'Swift', 'Bash', 'Matlab', 'Other Programming languages'],
    'Visualization Tool': ['Matplotlib', 'Seaborn', 'Plotly', 'Ggplot', 'Shiny', 'D3 js', 'Altair', 'Bokeh', 'Geoplotlib', 'Leaflet', 'Other Visualization tools'],
    'BI Tool': ['Power BI', 'Tableau', 'Qlik', 'Looker', 'Salesforce(Visuals)', 'Domo', 'TIBCO Spotfire', 'Alteryx', 'Sisense', 'Sap Analytics', 'Other Bi tools']}

melted_list = []

for question, cols in question_blocks.items():
    df_melt = df.melt(
        id_vars=id_vars,
        value_vars=cols,
        var_name='Option',   # Column name of the tool/language
        value_name='Answer'  # Value chosen
    )
    df_melt['Question'] = question
    df_melt = df_melt.dropna(subset=['Answer'])  # only selected options
    df_melt['Selected'] = 1  # binary indicator
    melted_list.append(df_melt)

DF_summary = pd.concat(melted_list, ignore_index=True)
```

```python
Summary = DF_summary.groupby(['Question', 'Option', 'Gender', 'Education Level']).agg(
    Total_Respondents=('Respondent_id','nunique')
).reset_index()
```

#### Export ready for Excel
